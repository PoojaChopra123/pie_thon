```python
def hourly(x):
    if x['H']==7:
        return '07-7am to 8am'
    elif x['H']==8:
        return '08-8am to 9am'
    elif x['H']==9:
        return '09-9am to 10am'
    elif x['H']==10:
        return '10-10am to 11am'
    elif x['H']==11:
        return '11-11am to 12pm'
    elif x['H']==12:
        return '12-12pm to 1pm'
    elif x['H']==13:
        return '13-1pm to 2pm'
    elif x['H']==14:
        return '14-2pm to 3pm'
    elif x['H']==15:
        return '15-3pm to 4pm'
    elif x['H']==16:
        return '16-4pm to 5pm'
    elif x['H']==17:
        return '17-5pm to 6pm'
    elif x['H']==18:
        return '18-6pm to 7pm'
    else:
        return '19-7pm to 8pm'
```

```python
df['H']=df.apply(hourly,axis=1)
```

```python
df.head()
```

```python
def daypart(x):
    if (
        x['H']=='07-7am to 8am'
    or
        x['H']=='07-7am to 8am'
    or
        x['H']=='08-8am to 9am'
    or
        x['H']=='09-9am to 10am'
    or
        x['H']=='10-10am to 11am'
    or 
        x['H']=='11-11am to 12pm'):
        
        return '01-Morning'

    elif ( 
        x['H']=='12-12pm to 1pm'
    or
        x['H']=='13-1pm to 2pm'
    or
        x['H']=='14-2pm to 3pm'
    or
        x['H']=='15-3pm to 4pm'):
        
        return '02-Afternoon'
    
    elif (    
        x['H']=='16-4pm to 5pm'
    or 
        x['H']=='17-5pm to 6pm'):
        
        return '03-Early Evening'
    
    else:
        return '04-Evening'
```

```python
df['DAY_PART']=df.apply(daypart,axis=1)
```

```python
df.head()
```

```python
def seasonal(x):
    
    if (
        x['MonthN']==12.0
    or 
        x['MonthN']==11.0
    or
        x['MonthN']==1.0
    or
        x['MonthN']==2.0):
        
        return 'Cold Season'
    
    elif (
        x['MonthN']==3.0
    or 
        x['MonthN']==4.0
    or
        x['MonthN']==5.0
    or
        x['MonthN']==9.0
    or 
        x['MonthN']==10.0):
        
        return 'Cool Season'
    
    else:
        return 'Hot Season'
```

```python
df['SEASONAL']=df.apply(seasonal,axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1)
```

```python

# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
```

```python
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
```

```python
# Train the model on training data
rf.fit(X_train, y_train);
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(X_test)
# Calculate the absolute errors
errors = abs(predictions - y_test)
# Print out the mean absolute error (mae)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
print(errors)
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (errors / y_test)
# Calculate and display accuracy
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```
