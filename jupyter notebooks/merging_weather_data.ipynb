```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
weather_file = pd.read_csv('weatherresults.csv')
```

```python
weather_file['Formatted_Date_1'] = pd.to_datetime(weather_file['Formatted_Date'])
weather_file.head()
```

```python
data_file = pd.read_excel('MasterData_Wilson.xlsx')
```

```python
data_file.head()
```

```python
merged_file = pd.merge(weather_file, data_file, left_on='Formatted_Date_1', right_on='date')
merged_file.head()
```

```python
del merged_file['Unnamed: 0']
del merged_file['Formatted_Date']
del merged_file['Formatted_Date_1']
del merged_file['Date']
del merged_file['Unnamed: 29']
del merged_file['Unnamed: 30']
merged_file.head()
```

```python
merged_file['Total Due'] = pd.to_numeric(merged_file['Total Due'], errors='coerce')
date_group = merged_file.groupby(['date'])['Total Due'].sum().to_frame(name = 'count').reset_index()
merged_file1 = merged_file.merge(date_group)
merged_file1.head()
```

```python
#plot total value per day depending on the weather
ax1 = sns.boxplot(x='Summary',y='count',data=merged_file1)
plt.xticks(rotation=90)
#too many categories, I am going to group categories together and we need a percentage 
```

```python
#merging weather into 4 categories
merged_file1["Summary"] = merged_file1["Summary"].replace(
    {"Drizzle": "Cloudy", "Partly Cloudy": "Cloudy", "Mostly Cloudy": "Cloudy","Foggy": "Cloudy","Humid": "Cloudy","Humid and Partly Cloudy": "Cloudy","Humid and Mostly Cloudy": "Cloudy","Clear": "Sunny","Overcast": "Sunny","Rain": "Rainy","Heavy Rain": "Rainy","Light Rain": "Rainy","Flurries": "Rainy", "Light Snow": "Snowy"})

ax2 = sns.boxplot(x='Summary',y='count',data=merged_file1)
plt.xticks(rotation=90)
```

```python
#groupby day of the week and weather, total sum / number of days in each category
day_group = merged_file1.groupby(['WeekD','Summary'])['count'].sum().rename("count")
number_of_days = merged_file1.groupby(['WeekD','Summary'])['count'].nunique()

day_group_normalized = day_group / number_of_days
day_group_normalized_df = day_group_normalized.to_frame()

Sum_perday = day_group_normalized.plot(kind='bar')
Sum_perday
```

```python
#Mean amount for each category
day_group_mean = merged_file1.groupby(['WeekD','Summary'])['count'].mean().to_frame(name='mean').reset_index()
merged_file2 = merged_file1.merge(day_group_mean)
merged_file2.head()

#split each day
Monday_df = merged_file2[merged_file2['WeekD'] == 'Mon']
Tuesday_df = merged_file2[merged_file2['WeekD'] == 'Tue']
Wednesday_df = merged_file2[merged_file2['WeekD'] == 'Wed']
Thursday_df = merged_file2[merged_file2['WeekD'] == 'Thu']
Friday_df = merged_file2[merged_file2['WeekD'] == 'Fri']
Saturday_df = merged_file2[merged_file2['WeekD'] == 'Sat']
Sunday_df = merged_file2[merged_file2['WeekD'] == 'Sun']

ax2 = sns.boxplot(x='Summary',y='count',data=Saturday_df)
```

```python
#Multiple plots side by side, order by weather category
fig, ax =plt.subplots(1,4,sharex='col', sharey='row',figsize=(18, 6))
sns.boxplot(x='Summary',y='count',data=Monday_df,order = ['Sunny','Cloudy','Rainy'], ax=ax[0]).set_title('Monday', fontsize=16)
sns.boxplot(x='Summary',y='count',data=Tuesday_df, order = ['Sunny','Cloudy','Rainy'],ax=ax[1]).set_title('Tuesday', fontsize=16)
sns.boxplot(x='Summary',y='count',data=Wednesday_df, order = ['Sunny','Cloudy','Rainy'],ax=ax[2]).set_title('Wednesday', fontsize=16)
sns.boxplot(x='Summary',y='count',data=Thursday_df, order = ['Sunny','Cloudy','Rainy'],ax=ax[3]).set_title('Thursday', fontsize=16)
plt.ylim(0,6000)
fig.show()
fig.savefig('weatherinfluence1.png')
```

```python
fig, ax =plt.subplots(1,3,sharex='col', sharey='row',figsize=(18, 6))
sns.boxplot(x='Summary',y='count',data=Friday_df, order = ['Sunny','Cloudy','Rainy'],ax=ax[0]).set_title('Friday', fontsize=16)
sns.boxplot(x='Summary',y='count',data=Saturday_df, order = ['Sunny','Cloudy','Rainy'],ax=ax[1]).set_title('Saturday', fontsize=16)
sns.boxplot(x='Summary',y='count',data=Sunday_df, order = ['Sunny','Cloudy','Rainy'],ax=ax[2]).set_title('Sunday', fontsize=16)
plt.ylim(0,6000)
fig.show()
fig.savefig('weatherinfluence2.png')
```
